```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('house-prices-advanced-regression-techniques\\train.csv')
```

```python
print(dataset.shape)
dataset.head()
```

# Missing_values

```python
dataset.info()
```

```python
featr_na=[featr for featr in dataset.columns if dataset[featr].isnull().sum()>1]
for featr in featr_na:
    print(featr,np.round(dataset[featr].isnull().mean()*100,4),'% missing values')
print(len(featr_na))
```

```python
for featr in featr_na:
    data=dataset.copy()
    data[featr]=np.where(data[featr].isnull(),1,0)
    data.groupby(featr)['SalePrice'].median().plot.bar()
    plt.title(featr)
    plt.show()
```

There is relation between the missing values and the dependent value sot it need to replaced with some meaningfull value

# Numerical_Variables

```python
num_featr=[featr for featr in dataset.columns if dataset[featr].dtypes !='O']
print('Number of numerical variables : {}'.format(len(num_featr)))
dataset[num_featr].head()
```

# Year_variables

```python
yr=['year','Year','YEAR','YR','yr','Yr']
yr_featr=[featr for featr in num_featr if any( YR in featr for YR in yr)]
yr_featr
```

```python
for featr in yr_featr:
    print(featr,dataset[featr].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.title('yrsold vs salesprice')
plt.ylabel('median_salesprice')
plt.show()
```

```python
for featr in yr_featr:
    if featr !='YrSold':
        data=dataset.copy()
        data[featr]=data['YrSold']-data[featr]
        plt.scatter(data[featr],data['SalePrice'])
        plt.xlabel(featr)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
descrt_featr=[featr for featr in num_featr if len(dataset[featr].unique())<25 and featr not in yr_featr+['Id']]
print('Discrete variable count: {}'.format(len(descrt_featr)))
```

```python
descrt_featr
```

```python
dataset[descrt_featr].head()
```

```python
for featr in descrt_featr:
    data=dataset.copy()
    data.groupby(featr)['SalePrice'].median().plot.bar()
    plt.title(featr)
    plt.ylabel('Sales price')
    plt.show()
```

# ContinuousFeature

```python
cont_featr=[featr for featr in num_featr if featr not in descrt_featr+yr_featr+['Id']]
print('Continuous variable count: {}'.format(len(cont_featr)))
cont_featr
```

```python
for featr in cont_featr:
    data=dataset.copy()
    data[featr].hist(bins=25)
    plt.title(featr)
    plt.ylabel('Count')
    plt.show()
```

```python
for featr in cont_featr:
    data=dataset.copy()
    if 0 in data[featr].unique():
        pass
    else:
        data[featr]=np.log(data[featr])
        plt.scatter(data[featr],data['SalePrice'])
        plt.title(featr)
        plt.ylabel('Sale Price')
        plt.show()
```

### outliers

```python
for featr in cont_featr:
    data=dataset.copy()
    if 0 in data[featr].unique():
        pass
    else:
        data[featr]=np.log(data[featr])
        sb.boxplot(y=data[featr])
        plt.show()
        
```

### Categorical Varables

```python
catg_featr=[featr for featr in dataset.columns if dataset[featr].dtype =='O']
print('Number of categorical variables : {}'.format(len(catg_featr)))
dataset[catg_featr].head()
```

```python
for featr in catg_featr:
    print('{} has {} number of categories.'.format(featr,len(dataset[featr].unique())))
```

```python
for featr in catg_featr:
    data=dataset.copy()
    data.groupby(featr)['SalePrice'].median().plot.bar()
    plt.title(featr)
    plt.ylabel('Sales price')
    plt.show()
```
