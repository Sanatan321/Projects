# missing values

```python
catg_nan=[featr for featr in dataset.columns if dataset[featr].isnull().sum()>1 and dataset[featr].dtypes =='O']
for featr in catg_nan:
    print(featr,np.round(dataset[featr].isnull().mean()*100,4),'% missing values')
```

```python
dataset.drop(['Alley'],axis=1,inplace=True)
dataset.drop(['PoolQC'],axis=1,inplace=True)
dataset.drop(['Fence'],axis=1,inplace=True)
dataset.drop(['MiscFeature'],axis=1,inplace=True)
dataset.drop(['FireplaceQu'], axis = 1, inplace = True)

data_test.drop(['Alley'],axis=1,inplace=True)
data_test.drop(['PoolQC'],axis=1,inplace=True)
data_test.drop(['Fence'],axis=1,inplace=True)
data_test.drop(['MiscFeature'],axis=1,inplace=True)
data_test.drop(['FireplaceQu'], axis = 1, inplace = True)
```

```python
catg_nan=[featr for featr in dataset.columns if dataset[featr].isnull().sum()>1 and dataset[featr].dtypes =='O']
```

```python
for featr in catg_nan:
    mode=dataset[featr].mode()[0]
    dataset[featr].fillna(mode,inplace=True)
```

```python
for featr in catg_nan:
    mode=data_test[featr].mode()[0]
    data_test[featr].fillna(mode,inplace=True)
```

```python
for featr in num_nan:
    median=dataset[featr].median()
    dataset[featr].fillna(median,inplace=True)
```

## Rare Categorical variables

```python
catg_featr=[featr for featr in dataset.columns if dataset[featr].dtype =='O']
```

```python
for featr in catg_featr:
    temp=dataset.groupby(featr)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[featr]=np.where(dataset[featr].isin(temp_df),dataset[featr],'Rare')
```

```python
for featr in catg_featr:
    temp=data_test.groupby(featr)['SalePrice'].count()/len(data_test)
    temp_df=temp[temp>0.01].index
    data_test[featr]=np.where(data_test[featr].isin(temp_df),data_test[featr],'Rare')
```

```python
dataset.head()
```

```python
for featr in catg_featr:
    print('{} has {} number of categories.'.format(featr,len(dataset[featr].unique())))
```

```python
df=pd.concat([dataset,data_test],axis=0)
df.shape
```

```python
df2 = pd.get_dummies(df, drop_first=True)
```

```python
data_df=df2.loc[:,~df2.columns.duplicated()]
```

## Feature Scaling

```python
featr_scale=[featr for featr in data_df.columns if featr not in['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
data_df[featr_scale]=scaler.fit_transform(data_df[featr_scale])
```

```python
columns=np.full((corr.shape[0],),True,dtype=bool)
for i in range(corr.shape[0]):
    for j in range(i+1,corr.shape[0]):
        if corr.iloc[i,j]>0.9:
            if columns[j]:
                columns[j]=False
```
